### 1. Imagine the year is 1950, and you are tasked with designing Folsom Reservoir. The only data at your disposal is a record of annual streamflows on the American River from 1901-1949 (see hw2-folsom-historical.csv). Use the sequent peak method to find the reservoir size that will provide a firm yield of 1500 TAF/yr.

```python
import numpy as np 
import matplotlib.pyplot as plt
from utility import simulate
# sequent peak example from LVB Table 11.2

Q = np.loadtxt('D:/UC DAVIS/ECI 273/hw2/hw2-folsom-historical.csv', delimiter=',', 
                skiprows=1, usecols=[1])
Q = Q/1000
T = len(Q)
R = 1500*np.ones(T)

def sequent_peak(R, Q):
  # accepts inflow and outflow arrays
  # returns sequent peak reservoir capacity
    assert len(R) == len(Q), 'R and Q must be the same length'
    
    T = len(Q)
    K = np.zeros(T+1)

    for t in range(T):
        K[t+1] = max(R[t] - Q[t] + K[t], 0)

    return np.max(K)


Kmax = sequent_peak(R,Q)
print("The reservoir size is", round(Kmax,2))

```

### Using the reservoir size you found in (1), run a simulation to find the reliability of delivering 1500 TAF/yr over the entire paleo record. Start the simulation with the reservoir full. What does this tell you about the importance of record length in the reservoir sizing process?

```python
Q = np.loadtxt('D:/UC DAVIS/ECI 273/hw2/hw2-folsom-paleo.csv', delimiter=',', 
                skiprows=1, usecols=[1])

Q = Q/1000
T = len(Q)
R = 1500*np.ones(T)
K = Kmax
D = 1500

S = np.zeros(T)
R = np.zeros(T)
shortage = np.zeros(T)

S[0] = K # start simulation full
R[0] = D # first day meets demand, THIS IS THE OUTFLOW
met_demand = 1 # counter

for t in range(1,T):
    # new storage: mass balance, max value is K
    S[t] = min(S[t-1] + Q[t-1] - R[t-1], K)
    
    # release is based on demand
    if S[t] + Q[t] > D:
        R[t] = D
        met_demand = met_demand + 1
    else:
        R[t] = S[t] + Q[t]

        shortage[t] = D-R[t]

reliability = met_demand / T
print('The reliability is', round(reliability,3)*100, '%')
```

Initially, I figured that using paleo data for the sizing process would be beneficial as it provides a longer range which could give a more accurate answer for the reservoir sizing. However after calculating the reliability for the streamflow, the difference between using the paleo data and the historical data is only about 1%. Hence in this scenario, the extra data gathered from tree rings might not be the most important as it most likely did not add much more information regarding the the general pattern of inflows.

### 4) Use the paleo record and your simulation model to plot the reliability-capacity curve for a fixed yield of 1500 TAF/yr. Start all simulations with the reservoir full. On the graph, mark the size of the actual reservoir (975 TAF).

```python
Q = np.loadtxt('D:/UC DAVIS/ECI 273/hw2/hw2-folsom-paleo.csv', delimiter=',', 
                skiprows=1, usecols=[1])

Q = Q/1000
T = len(Q)
R = 1500*np.ones(T)
K = Kmax
D = 1500

S = np.zeros(T)
R = np.zeros(T)
shortage = np.zeros(T)

S[0] = K # start simulation full
R[0] = D # first day meets demand, THIS IS THE OUTFLOW
met_demand = 1 # counter

for t in range(1,T):
    # new storage: mass balance, max value is K
    S[t] = min(S[t-1] + Q[t-1] - R[t-1], K)
    
    # release is based on demand
    if S[t] + Q[t] > D:
        R[t] = D
        met_demand = met_demand + 1
    else:
        R[t] = S[t] + Q[t]

        shortage[t] = D-R[t]

K = np.arange(0,3000,100) # make an array of storage values
reliability = np.zeros(len(K)) # we'll calculate these for each K

for i in range(len(K)):
    reliability[i] = simulate(K[i], D, Q)
    
plt.plot(K, reliability)
plt.plot(975, simulate(975, D, Q), 'r+')
plt.xlabel('Capacity (TAF)')
plt.ylabel('Reliability')
plt.title('Reliability-Capacity Curve for 1500 TAF/m Yield')
plt.show()
```

### 5. Create a heatmap where the x-axis is reservoir capacity, the y-axis is yield, and the color scale is reliability.

```python
K = np.arange(0,3000,100) # make an array of storage values
D = np.arange(0,1500,50)
reliability = np.zeros((int(len(K)),int(len(D)))) # we'll calculate these for each K
for i in range(len(K)):
    for j in range(len(D)):
        reliability[i,j] = simulate(K[i], D[j], Q)
        
plt.imshow(reliability.T, interpolation = 'None', origin = 'lower', extent = [0,3000,0,1500], aspect = 2, cmap=plt.cm.hot)
plt.colorbar()
plt.xlabel('Reservor Capacity (TAF)')
plt.ylabel('Yield (TAF)')
plt.title('Image Plot of Reliability')
plt.show
```
